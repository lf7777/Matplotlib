
# Matplotlib 学习


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 准备数据

```python
data = pd.DataFrame(index=[0,1,2,3,4],columns=['DATE','VALUE'])
data['DATE']=['1948-01-01','1948-02-01','1948-03-01','1948-04-01','1948-05-01']
data['DATE']=pd.to_datetime(data['DATE'])
data['VALUE']=[3.4,3.8,4.0,3.9,3.5]
data['MONTH']=data['DATE'].dt.month
data
```

```python
# 判断数据是否有空值
nan = data.isnull()
nanNum = nan.sum()
nanNum
```

# 创建空图形

```python
t = np.array(['飞机','轮船','汽车','自行车','走路'])
t
```

```python
# 折线图 plot
plt.plot(t,[3,1,2,5,9])
```

```python
# 散点图 scatter
plt.scatter(t,[3,1,2,5,9])
plt.show()
```

```python
# 柱状图 bar
plt.bar(t,[3,1,2,5,9])
plt.show()
```

```python
# 条形图
plt.barh(t,[3,1,2,5,9],color='r')
plt.show()
```

```python
# 直方图 hist
plt.hist([1,2,3,3,4,4,5,5,5,5,5,6,7])
plt.show()
```

```python
#饼图 pie
plt.figure(figsize=(8,8))
plt.pie([2.2,0.2,0.56,0.3,0.8],
        labels=['china','russia','india','america','japan'],
        autopct='%1.2f%%',
        shadow=True,
        explode=(0.05,0.03,0.04,0.05,0.05),
        startangle=-100
       )
plt.show()
```

```python
# 箱线图 boxplot
plt.boxplot([15,5,9,22,4,-5,45,-22])
plt.show()
```

```python
# 热力图 imshow()
data = np.random.randint(1,100,60).reshape(2,30)
data
plt.imshow(data,cmap='gray',origin='lower',interpolation='lanczos')
plt.show()
```
